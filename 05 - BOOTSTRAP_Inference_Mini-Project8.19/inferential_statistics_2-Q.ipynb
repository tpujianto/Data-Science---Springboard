__Q:__ Use bootstrap sampling to estimate the same 95% confidence interval lower limit as before.

__A:__

```python
np.random.seed(47)
N_rep = 10000

# use numpy to generate bootstrap replicate
bootstrapped = np.empty(N_rep)
for i in range(N_rep):
    bootstrapped[i] = np.mean(np.random.choice(med_charges['charges'], size=len(med_charges['charges'])))

# find the 95% lower Confidence Interval
CI = np.quantile(bootstrapped, .05)
print("The 95% Confidence Interval lower limit from the bootstrapped sample is: {:.2f}".format(CI))
```

If you performed 10000 replicates immediately after setting the random seed to 47, you should get the value 12724 here, which compares very well with the value 12725 obtained using the _t_-distribution confidence interval previously. It is a most pleasant result to see the predictions of classical frequentist theory match with results that are now possible through the number-crunching ability of computers.

__Q:__ Calculate the 95% confidence interval for the difference between the standard deviations of insurance and non-insurance claim charges (insured - non-insured). Calculate the differences over 10000 replicates. Plot the histogram of values and mark the locations of the percentiles. State the null and alternative hypothesis and comment on whether you would retain or reject the null hypothesis in this case and why.

__A:__<br>
$H_0$: The charges of insured and uninsured are equal variance. <br>
$H_\alpha$: The charges of insured and uninsured are not equal variance. <br>




```python
# creating function to draw bootstrap replicates
def draw_bs_reps(data, func, size=1):
    """Draw bootstrap replicates."""

    # Initialize array of replicates: bs_replicates
    bs_replicates = np.empty(size)

    # Generate replicates
    for i in range(size):
        bs_samples = np.random.choice(data, size=len(data))
        bs_replicates[i] = func(bs_samples)

    return bs_replicates
```

```python
# extract the data for each, insured and uninsured charges
insured = med_charges.loc[med_charges['insuranceclaim'] == 1]['charges']
uninsured = med_charges.loc[med_charges['insuranceclaim'] == 0]['charges']
```

```python
# utilizing the draw_bs_replicates to generate 10,000 bootstrap replicates for the 2 groups
boot_insured_sd = draw_bs_reps(insured, np.std, size=10000)
boot_uninsured_sd = draw_bs_reps(uninsured, np.std, size=10000)

# find the differences
boot_differences_sd = boot_insured_sd - boot_uninsured_sd
```

```python
# find 2.5% percentile and 9.5% percentile
lower, upper = np.quantile(boot_differences_sd, [0.025, 0.975])
```

```python
_ = plt.hist(boot_differences_sd, bins=30)
_ = plt.xlabel('Difference of Standard Deviation')
_ = plt.ylabel('USD')
_ = plt.title('Insured and Uninsured Charges')
_ = plt.axvline(lower, color='r', linestyle='--')
_ = plt.axvline(upper, color='m', linestyle='--')
```

```python
print("We would fail to reject the null hypothesis if we are confident there is not significant difference or around 0 between insured and uninsured standard deviation.")
print("Since we are 95% confident the difference of standard deviation is between {:.2f} and {:.2f}. We reject the null hypothesis.".format(lower, upper))
```

## Confidence interval and p-value

The confidence interval above is often a useful quantity to estimate. If we wish to limit our expected probability of making a Type I error (where we wrongly reject the null hypothesis, and it is, instead, true) to $\alpha$, the associated confidence interval is our estimate of the interval within which we expect the true population value to be found $100\times(1 - \alpha)$% of the time we do this test. In the above we performed bootstrap replicates to estimate the interval and reject the null hypothesis if this interval did not contain zero. You will sometimes see such an interval reported in the output of statistical functions.

The partner of the confidence interval is the p-value. The p-value and the confidence interval are linked through our choice of $\alpha$. The p-value tells us how likely it is, under the null hypothesis, to get an outcome at least as extreme as what was observed. If this fails to reach the level of our _pre-specified_ $\alpha$, we decide the null hypothesis is sufficiently unlikely to be true and thus reject it. To calculate this p-value via the bootstrap, we have to put ourselves in a position where we are simulating the null hypothesis being true and then calculate the fraction of times we observe a result at least as extreme as that actually observed.

Remember how, previously, you used the _t_-test to calculate the p-value for the observed difference between the means of insured and non-insured medical cases. We're now going to repeat this, this time using the bootstrap approach.

__Q:__ Perform a bootstrapped hypothesis test at the 5% significance level ($\alpha = 0.05$) to calculate the p-value of the observed difference between insurance and non-insurance charges, state your null and alternative hypotheses and whether you retain or reject the null hypothesis for the given significance level.

__A:__<br>
$H_0$: There is no difference in charges between people with insurance and people without insurance <br>
$H_\alpha$: There is a difference in the amount charged between the two groups <br>

```python
# from the null hypothesis, our first assumption is the mean of charges between the 2 groups are the same, thus we need to shift the mean of both groups 
merge_mean = np.mean(np.concatenate((insured, uninsured)))
insured_s = insured - np.mean(insured) + merge_mean
uninsured_s = uninsured - np.mean(uninsured) + merge_mean
```

```python
# utilizing the draw_bs_replicates to generate 10,000 bootstrap replicates for the 2 shifted groups, this time we're interested in their mean
boot_insured_mean = draw_bs_reps(insured_s, np.mean, size=10000)
boot_uninsured_mean = draw_bs_reps(uninsured_s, np.mean, size=10000)

# find the difference
boot_differences_mean = boot_insured_mean - boot_uninsured_mean
```

```python
# calculating p-value
obbserved_mean_diff = np.mean(insured) - np.mean(uninsured)
p_val = np.sum(boot_differences_mean >= obbserved_mean_diff) / len(boot_differences_mean)
p_val
```

```python
if p_val < 0.05:
    print("The p-value : {} is smaller than alpha: 0.05. Thus, we reject the null hypothesis.".format(p_val))
else:
    print("The p-value : {} is larger than alpha: 0.05. Thus, we fail to reject the null hypothesis.".format(p_val))
```
